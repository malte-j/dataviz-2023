```python
import pandas as pd

df = pd.read_csv('../archive/csv/game.csv')

print(df.columns)
```

```python
import matplotlib.pyplot as plt

# group by game_date year
df.groupby(df['game_date'].str[:4]).count()['game_id'].plot(kind='bar')

# increase plot size
plt.rcParams['figure.figsize'] = [18, 10]


plt.show()
```
