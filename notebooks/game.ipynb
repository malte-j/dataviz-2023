```python
# import duckdb
import pandas as pd
# No need to import duckdb_engine
#  jupysql will auto-detect the driver needed based on the connection string!

# Import jupysql Jupyter extension to create SQL cells
%reload_ext sql
%load_ext sql
%config SqlMagic.autopandas = True
%config SqlMagic.feedback = False
%config SqlMagic.displaycon = False
%sql duckdb:///:default:
# Plot display settings
import matplotlib.pyplot as plt
import matplotlib

plt.rcParams['figure.dpi'] = 300
matplotlib.rcParams['font.family'] = ['Inter', 'sans-serif']

pd.set_option('display.max_colwidth', 3000)
pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", None)
```

```python
%sql away_wins << SELECT season_id, team_abbreviation_away, count(*) as count FROM read_csv_auto('../archive/csv/game.csv') WHERE wl_away = 'W' GROUP BY season_id, team_abbreviation_away
away_wins: pd.DataFrame

away_wins['season'] = away_wins['season_id'].astype(str).str[1:]
```

```python
%sql home_wins << SELECT season_id, team_abbreviation_home, count(*) as count FROM read_csv_auto('../archive/csv/game.csv') WHERE wl_home = 'W' GROUP BY season_id, team_abbreviation_home
home_wins: pd.DataFrame

home_wins['season'] = home_wins['season_id'].astype(str).str[1:]
```

```python
%sql team_abbreviatons_to_name << SELECT team_abbreviation_home as team_abbreviation, team_name_home as team_name FROM read_csv_auto('../archive/csv/game.csv') GROUP BY team_abbreviation_home, team_name_home;


team_abbreviatons_to_name: pd.DataFrame
# remove duplicates
team_abbreviatons_to_name = team_abbreviatons_to_name.drop_duplicates(subset=['team_abbreviation'])
team_abbreviatons_to_name
```

```python
# read image urls from imgs.csv
imgs = pd.read_csv('./data/imgs.csv')
# drop name,active,image
imgs = imgs.drop(['name', 'active', 'url'], axis=1)
imgs = pd.merge(imgs, pd.read_csv('./data/colors.csv'), on='id')
```

```python
# merge home and away wins and add count
wins = pd.merge(home_wins, away_wins, how='inner', left_on=['season', 'team_abbreviation_home'], right_on=['season', 'team_abbreviation_away'])
wins['count'] = wins['count_x'] + wins['count_y']

#rename team_abbreviation_home to team_abbreviation
wins = wins.rename(columns={'team_abbreviation_home': 'team_abbreviation'})

# drop unnecessary columns
wins = wins.drop(['season_id_x', 'season_id_y', 'team_abbreviation_away',  'count_x', 'count_y'], axis=1)

# join with team_abbreviatons_to_name
# wins = pd.merge(wins, team_abbreviatons_to_name, how='left', left_on=['team_abbreviation'], right_on=['team_abbreviation'])


wins
```

```python
# change, so that each season is a column

wins_pvt = wins.pivot(index='team_abbreviation', columns='season', values='count')

# index to column
wins_pvt = wins_pvt.reset_index()

# copy team_abbreviation to id
# wins_pvt['id'] = wins_pvt.index

# add imgs
wins_pvt = pd.merge(wins_pvt, imgs, how='left', left_on=['team_abbreviation'], right_on=['id'], copy=True)

# add team name
wins_pvt = pd.merge(wins_pvt, team_abbreviatons_to_name, how='left', left_on=['team_abbreviation'], right_on=['team_abbreviation'], copy=True)

# drop unnecessary columns
wins_pvt = wins_pvt.drop(['team_abbreviation', 'id'], axis=1)

wins_pvt
wins_pvt.to_csv('../exports/wins_pvt.csv', index=False)
```
